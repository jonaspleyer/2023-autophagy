```python
import numpy as np
import matplotlib.pyplot as plt

def potential(sigma, cutoff):
    mod = sigma < cutoff
    s = -(1/sigma**2 - 1/(sigma**4))
    bound = 4 + 1/sigma
    return mod* np.min(np.array([s, bound]), axis=0)

fig, ax = plt.subplots()

cutoff = 1.4
x = np.linspace(0.1, 1.5 * cutoff, 200)
y = potential(x, cutoff)

ax.set_title("Relative Interaction Strength")
ax.set_xlabel('Dimnensionless Radial Coordinate $\\sigma$')
ax.set_ylabel('Strength $V/U_i$')
ax.plot(x, y, color="black")
ax.plot(x, 0*x, color="grey", alpha=0.5, linestyle="--")
ax.set_ylim([None,10])
fig.savefig("interaciton_potential.png")
```
